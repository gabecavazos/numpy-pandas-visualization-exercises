```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
#load iris
iris = sns.load_dataset('iris')
```

```python
iris.columns
```

```python
iris.info()
```

```python
#distribution of petal length
sns.displot(
data=iris,
x='petal_length')
```

```python
#2 is there correlation between petal length and width?
sns.scatterplot(
data=iris,
x='petal_length',
y='petal_width')
plt.show()
#yes
```

```python
#3 Would it be reasonable to predict species based on sepal width and sepal length? For this, 
#you'll visualize two numeric columns through the lense of a categorical column.
sns.relplot(
data=iris,
x='sepal_width',
y='sepal_length',
hue='species')
plt.show()
#it would be possible to separate setosas from versicolor and verginica 
#but the other two would be too inacurate unless petal length is over 7
```

```python
#4 which features would be best to predict species?
sns.relplot(
data=iris,
x='sepal_width',
y='petal_length',
hue='species')
plt.show()
#sepal width and petal length
```

```python
#Using the chipotle dataset from the previous exercise, 
#create a bar chart that shows the 4 most popular items and the revenue produced by each.
from env import user,password,host
def get_db_url(user,password,host,database_name):
    url =  f'mysql+pymysql://{user}:{password}@{host}/{database_name}'
    return url
```

```python
url = get_db_url(user,password,host,'chipotle')
```

```python
pd.read_sql('''
show table
''', url)
```

```python
chipotle = pd.read_sql('''
SELECT *
FROM orders''', url)
```

```python
chipotle.info()
```

```python
chipotle.head()
```

```python
chipotle['item_price'] = chipotle.item_price.str.strip('$').astype(float)
```

```python
chipotle['revenue']= chipotle.item_price * chipotle.quantity
```

```python
popular_items = ['Chicken Bowl','Chicken Burrito','Chips and Guacamole','Steak Burrito']
chipotle['is_popular_item'] =chipotle.item_name.isin(popular_items)
```

```python
sns.barplot(
    data=chipotle[chipotle.is_popular_item == True],
    x='item_name',
    y='revenue')
plt.show()
```
